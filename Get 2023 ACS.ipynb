```python
#
# Collect the ACS 5 Year estimates for the variables in the myvars.csv
#

import pandas as pd
import censusdata

myvars = pd.read_csv('acsvars.csv')

# Pull all block groups for one state at a time

states = censusdata.geographies(censusdata.censusgeo([('state', '*')]), 'acs5', 2022)
statelist = []

for state in states.items():
    print("Downloading",state[1].geo[0][1],state[0])
    statelist.append(
        censusdata.download('acs5', 2023, censusdata.censusgeo([('state', state[1].geo[0][1]), ('county', '*'), ('block group', '*')]),
                                    ['GEO_ID']+list(myvars.acsformat))
    )
    
acs = pd.concat(statelist)
print('ACS collected')

# Now let's add the land area and water area from the TIGER Line publication.
tiger = pd.read_csv('../blockgroupcartographics/2023blockgroupcartographics.csv', dtype={
                        "STATEFP":str,
                        "COUNTYFP":str,
                        "TRACTCE":str,
                        "BLKGRPCE":str,
                        "GEOID":str,
                        "GEOIDFQ":str,
                        "NAMELSAD":str,
                        "MTFCC":str,
                        "FUNCSTAT":str,
                        "geometry":str,
                        "GEO_ID":str})

full = pd.merge(acs, tiger[['GEO_ID','ALAND', 'AWATER', 'INTPTLAT','INTPTLON',]], how='left', on='GEO_ID' )

full.to_csv('2023blockgroupdemographics.csv')

print("Merged dataset written")
```
